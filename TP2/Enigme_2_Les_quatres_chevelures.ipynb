```python
# Enigme : définir la couleur de notre chevelure, en ne voyant seulement que les chevelures des personnes devant nous
```

```python
# On importe les différents modules pour faire du quantique
```

```python
import numpy as np
```

```python
from qiskit import *
```

```python
# On créé un circuit de 8 éléments, nos 4 personnages, ainsi que les 4 raisonnements
```

```python
enigme = QuantumCircuit(8,8)
```

```python
# On instaure le caractère aléatoire sur les 4 premiers qubits des personnages afin de voir toutes les combinaisons au 
# même moment
```

```python
enigme.h(0)
enigme.h(1)
enigme.h(2)
enigme.h(3)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# On calcul, pour q0, le nombre de chevelure de couleur indigo afin de savoir s'il est pair ou impair
```

```python
enigme.cx(1,4)
enigme.cx(2,4)
enigme.cx(3,4)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# Les autres personnages prennent conscience du nombre pair ou impair de chevelure indigo, ce qui va permettre de 
# connaître la couleur de leur propre chevelure
```

```python
enigme.cx(4,5)
enigme.cx(4,6)
enigme.cx(4,7)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# q1 prend conscience de sa couleur et permet aux autres personnages devant lui de savoir s'il existe un nombre pair ou
# impair de chevelures indigo en fonction de sa réponse
```

```python
enigme.cx(2,5)
enigme.cx(3,5)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# Le raisonnement des deux derniers personnages change en fonction de la réponse de q1
```

```python
enigme.cx(5,6)
enigme.cx(5,7)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# q2 prend conscience de sa couleur et fournit l'information a q3
```

```python
enigme.cx(3,6)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# On prend conscience de l'information du 3ème personnage et ça nous permet de connaître forcément la couleur de notre
# chevelure
```

```python
enigme.cx(6,7)
enigme.barrier(0,1,2,3,4,5,6,7)
enigme.draw('mpl')
```

```python
# On mesure tous les qubits afin de connaître les probabilités de chaque situation
```

```python
enigme.measure_all(add_bits=False)
enigme.draw('mpl')
```
